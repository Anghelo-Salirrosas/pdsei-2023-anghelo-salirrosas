### Negativo de una imagen

* s = 255- f(r)

````python
r = cv2.imread(imgSrc)[..., ::-1]/255
s = 1 - r #negativo de una imagen normalizada
plt.imshow(s)
plt.show()
````

### Tranformacion Logarítmica
* s = c.log(1+r)

````python
img = cv2.imread(imgSrc, 0)
#Esto se llama reducir el rango dinamico
#Usado para estudio de espectros de Fourier
fft_2 = np.fft.fft2(img) #tranformada rapida de fourier para dos dimensiones, pero ta decentrado
fft_2_abs = np.abs(fft_2)#como son numeros complejos, trabajamos con el modulo
fft_2_abs = np.fft.fftshift(fft_2_abs)#centramos el espectro
fft_2_abs_log = np.log(fft_2_abs+1)#se aplica la transformacion logaritmica, este espectro es para que nosotros lo podamos ver
print(np.min(fft_2_abs), np.max(fft_2_abs))
plt.figure(figsize=(10,15))
plt.imshow(fft_2_abs_log, cmap='gray')
plt.show()
````

### Transformación arbitraria
* ####   Especificar una funcion de transformacion
cv2.LUT(img, lut)

````python
funcion_ejemplo=np.arange(256)
funcion_ejemplo[90:140]=200

img = cv2.imread(imgSrc,0)

img_out = np.empty_like(img)

for m in range(img.shape[0]):
    for n in range(img.shape[1]):
        img_out[m,n]=funcion_ejemplo[img[m,n]]

        
plt.subplots(2,1,figsize=(8,10))
plt.subplot(211)
plt.plot(funcion_ejemplo)
plt.grid(True)
plt.title("Funcion")
plt.subplot(212)
plt.imshow(img_out, cmap="gray", vmin=0, vmax=255)
plt.title("Imagen con funcion aplicada")
#Esta es la forma tradicional, pero mas rapido con la funcion LUP
````

````python
#Tambien se pueden hacer transformaciones arbitarias
#constrain strengh? alargamiento de histograma

r = np.zeros(256)

for pix in range(256):
    if pix < 50:
        r[pix] = 30/50*(pix-50) + 30
    if 50 <=pix < 100:
        r[pix] = 190/50*(pix-50) + 30
    if 100 <=pix:
        r[pix] = 35/155*(pix-100) + 220
        
plt.plot(r)
plt.show()
````

````python
img = cv2.imread(imgSrc,0)
s=cv2.LUT(img, r) #lo hace automaticamente opencv y lo optimiza
plt.figure(figsize=(8,10))
plt.imshow(s, cmap="gray")
plt.show()
````

#####  Normalizar CDF $ \frac{cdf*hist.max}{cdf.max} $

````python
#Los histogramas es como contar cuantas veces aparecen un gris en una imagen y obtener una distribucion estadistica
#M y N son las filas y columnas, distribucion de probabilidad que aparezca un pixel
````

````python
#Numpy
img=cv2.imread(imgSrc)[...,::-1] #darle vuelta a la imagen
img_gray=img[...,0]

hist,bins=np.histogram(img_gray,256,[0,256]) #img, #barritas, #intervalo

#los bins no vamos a usar

plt.figure(figsize=(16,8))
plt.subplot(121)
plt.plot(np.arange(256),hist)
plt.subplot(122)
plt.bar(np.arange(256),hist)

#si lo normalizamos, tendriamos recien PDF, distribucion de probabilidad xD
````

````python
#Numpy
````

````python
#matplotlib
#recibe una iamgen linealizada ravel() te aplana un vector, un array de 2d lo convierte en 1d
hist=plt.hist(img_gray.ravel(),256,[0,256])

#internamente se puede hacer la distribucion PDF
plt.figure()
hist2=plt.hist(img_gray.ravel(),256,[0,256], density=True) #normalizado :V

cdf = np.cumsum(hist2[0])
plt.figure()
plt.plot(cdf) #es una funcion acumulativa q se va ploteando

plt.figure()
#ahora para aplicarlo
cdf_norm=cdf*hist2[0].max()/cdf.max()
plt.plot(cdf_norm)
#entonces, con este ultimo podemos aplicarlo a la imagen para obtener con LUT con la funcion acumulativa
#a esa aplicaicon de usar esta para la imagen de entrada, se le llama ecualizacion de histograma
````

### Equalización de Histograma adaptativa
*clahe = createCLAHE() 
*clahe.apply(img) 

* [CLAHE Histogram Eqalization](https://www.geeksforgeeks.org/clahe-histogram-eqalization-opencv/)

````python
#podemos tener variaciones de iluminacion, este comando es propio de opencv para ecualizacion adaptativa

img=cv2.imread(imgSrc)[...,::-1]
img_gray=img[...,0]

clahe=cv2.createCLAHE(clipLimit=5)
final_img = clahe.apply(img_gray)+30

plt.imshow(final_img,cmap="gray", vmin=0, vmax=255)

#mayormente adaptativo es mejor, pero aumenta el calculo computacional, demora el timing, rendimiento: bajar la equalizacion
#procesamiento mas liviano
#ROBERTS ES MAS LIVIANO PERO MENOS PRECISION, ETC...

#LA ECUALIZACION MEJORA PORQUE DISTRIBUYE LOS PIXELES DE FORMA UNIFORME, AL MEJORAR CONTRASTE
#IMAGENES ENBLANQUECIDAS, ENNGRECIDAS, GRISACEAS, ETC, LAS MEJORA AL AUMENTAR EL CONTRASTE DE LA IMAGEN
````
