### **5.4. RESULTADO C**


#### Implementa el algoritmo de especificación de histograma de una imagen en escala de grises, cuyo procedimiento se resume en la figura 3.

````python
#Con OpenCV
imgSrcC='Imagenes/Lab3/rayosX.jpg'
imgC=cv2.imread(imgSrcC)
hist=cv2.calcHist([imgC],[0],None,[256],[0,256]) #[img], [canal], mascara, [bins], [rango de datos] || [1]:verde [2]:azul

print("Con OpenCV")
plt.figure(figsize=(15,10)) #(20,6)
plt.subplot(221)
plt.plot(hist) #y ahi esta la grafica
plt.title("Histograma con OpenCV")
plt.subplot(222)

plt.bar(np.arange(256),hist.reshape(1,256)[0]) #antes de graficarlo como barrita, hay que reconvertirlo
plt.title("Histograma con barras")

# si queremos ver las capas de colores
plt.subplot(223)
color=("r*-","gs-","b^-")
for i, col in enumerate(color):
    hist_rgb=cv2.calcHist([imgC],[i],None,[256],[0,256])
    plt.plot(hist, col)
plt.title("Histograma RGB")

plt.subplot(224)
#print(hist.max())
#for i in range(0,255):
#    print(hist[i][0])

histCV=np.zeros(256)
for i in range(0,255):
    histCV[i]=hist[i][0]*255/hist.max()

#print(histCV)
#print(histCV.max())

cdfCV_norm = np.cumsum(histCV)
plt.plot(cdfCV_norm)
plt.title("Distribución CDF normalizada Manual")

plt.show()

print("Con Matplotlib")

#Con Matplotlib
img=cv2.imread(imgSrcC)[...,::-1] #darle vuelta a la imagen
img_gray=img[...,0]
#recibe una iamgen linealizada ravel() te aplana un vector, un array de 2d lo convierte en 1d
plt.figure(figsize=(15,10))
plt.subplot(221)
hist=plt.hist(img_gray.ravel(),256,[0,256])
plt.title("Histograma con Matplotlib")

#internamente se puede hacer la distribucion PDF
plt.subplot(222)
hist2=plt.hist(img_gray.ravel(),256,[0,256], density=True) #normalizado
plt.title("Histograma Normalizado")

plt.subplot(223)
cdf = np.cumsum(hist2[0])
plt.plot(cdf) #es una funcion acumulativa q se va ploteando
plt.title("Distribucion CDF con plt")

#ahora para aplicarlo
plt.subplot(224)
cdf_norm=cdf*hist2[0].max()/cdf.max()
plt.plot(cdf_norm)
plt.title("Distribución CDF normalizada")

plt.show()

#entonces, con este ultimo podemos aplicarlo a la imagen para obtener con LUT con la funcion acumulativa
#a esa aplicaicon de usar esta para la imagen de entrada, se le llama ecualizacion de histograma
print("Resultados mediante Imagenes Obtenidas")
img = cv2.imread(imgSrcC,0)
s=cv2.LUT(img, cdf_norm) #lo hace automaticamente opencv y lo optimiza
s2=cv2.LUT(img,cdfCV_norm)
plt.figure(figsize=(16,10))
plt.subplot(221)
plt.imshow(img,cmap="gray")
plt.title("Imagen Original")
plt.subplot(222)
plt.imshow(s, cmap="gray")
plt.title("Imagen Con Ecualización de Matplotlib")
plt.subplot(223)
plt.imshow(s2, cmap="gray")
plt.title("Imagen Con Ecualización de OpenCV Manual")
plt.subplot(224)

#HASTA ACA SE HIZO ANTES LO ESCRITO ANTES
img=cv2.imread(imgSrcC)[...,::-1]
img_gray=img[...,0]

img_equ=cv2.equalizeHist(img_gray)
plt.imshow(img_equ, cmap="gray", vmin=0, vmax=255)
plt.title("Imagen Con Ecualización de OpenCV Comando")

plt.show()
````

### b. Pregunta 2
- **Defina una función que ingresando un rango de entrada (ej. 100, 120), se cree una función para expandir ese rango, en todo el rango dinámico de 0 a 255.**

````python
def expandir_rango(rango_entrada, valor_minimo=0, valor_maximo=255):
   
    min_entrada, max_entrada = rango_entrada # Valoresmínimo y máximo del rango de entrada
    
    pendiente = (valor_maximo - valor_minimo) / (max_entrada - min_entrada) #Se busca la forma de la recta para el mapeo
    termino_independiente = valor_minimo - (min_entrada * pendiente)
    
    def mapeo(valor):
        return int(pendiente * valor + termino_independiente) # Definimos la función de mapeo y expresamos la salida en int
                                                              # para ser mas eficientes, tambien se podria en double redondeo 2
    
    return mapeo

#VALORES DE ENTRADA PARA JUGAR: RANGO Y CANTIDAD
rango_entrada = (100, 120)
cantidad = 50

lista_entrada=np.random.uniform(rango_entrada[0],rango_entrada[1],cantidad) #Generamos una gama de valores dentro del rango
#no necesariamente en orden, para que funcione de forma general

# Agregamos los valores límite al principio y al final de la lista
lista_entrada = np.append(rango_entrada[0], lista_entrada) #conjunto de elementos a agregar, lista donde se va agregar
lista_entrada = np.insert(lista_entrada, len(lista_entrada), rango_entrada[1])


#print(lista_entrada)

#Utilizando con el rango por ejemplo
mapeo_expandido = expandir_rango(rango_entrada) #Definimos y preparamos la funcion

#Nota: Se determina el rango de entrada, dado que se pide ingresar un rango de entrada, si fuera mas general
#se deberá variar un poco, determinando un nuevo rango a partir del min y max de la lista de valores a mapear

#print(rango_entrada[0])

lista_mapeo_expandido=np.zeros(len(lista_entrada))

contador=0


#for valor_entrada in range(rango_entrada[0], rango_entrada[1]+1):

print("Valores de entrada vs Salida")
print()
for valor_entrada in lista_entrada:
    valor_salida = mapeo_expandido(valor_entrada) #Usamos la funcion
    lista_mapeo_expandido[contador]=valor_salida
    contador=contador+1
    print(f'Valor de entrada: {valor_entrada}, Valor expandido: {valor_salida}')

print()
print("También tenemos la lista de mapeo expandido de salida al rango de 0 a 255")
print(lista_mapeo_expandido)

#Nota: Otra posibilidad que planteamos fue restando el valor minimo y multiplicando por 255, para luego dividirlo 
#por la diferencia entre el rango de entrada
````

### c. Pregunta 3
- **Definir una función para aumentar o reducir el brillo de una imagen, según valor especificado.**

````python

def ajustar_brillo(imgSrc, factor_brillo):
    
    img = cv2.imread(imgSrc)[..., ::-1] / 255.0 #Cargamos la imagen

    img_ajustada = img ** factor_brillo # Aplicamos la transformación gamma para ajustar el brillo

    
    img_ajustada = np.clip(img_ajustada, 0, None) # Nos aseguramos de que los valores estén dentro del rango [0, 1] con np.clip()

    return img_ajustada

#VALORES DE ENTRADA:
imgSrc = "Imagenes/Lab3/gato_con_botas.jpg" #Introducimos la direccion
factor_brillo = 0.7  # Ajustamos el valor del brillo

imagen_ajustada = ajustar_brillo(imgSrc, factor_brillo)

#Imagen Original
img_original = cv2.imread(imgSRC)[..., ::-1]

# Se tiene como resultado
plt.figure(figsize=(15, 6))
plt.subplot(121)
plt.title("Imagen Original")
#plt.imshow(cv2.imread(imgSrc)[..., ::-1] / 255.0)
plt.imshow(img_original)

plt.subplot(122)
plt.title("Imagen con Ajuste de Brillo")
plt.imshow(imagen_ajustada)
plt.show()
````
